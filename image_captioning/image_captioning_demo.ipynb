```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df_captions = pd.read_csv('images_caption.csv', header=None)
df_captions.columns = ['image_name', 'caption']
df_captions
```

```python
IMAGES_DIR = 'images'
fig = plt.figure(figsize=(15, 25))
for i, row in df_captions.iterrows():
    image_name, caption = row[0], row[1]
    _=plt.subplot(3, 2, 1+i)
    _=plt.imshow(plt.imread(os.path.join(IMAGES_DIR, image_name)))
    _=plt.title(caption+'\n', size=20)
    _=plt.axis('off')
plt.tight_layout()
plt.show()
```
